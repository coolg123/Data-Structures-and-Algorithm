```python
# Working On Training Data First

import pandas as pd

df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')
```

```python
# showing train.csv
df_train.head()
```

```python
# testing data

df_test.head()
```

```python
df_train.isnull().sum()
```

```python
# replacing the null values for both train and test

df_train["Age"] = df_train["Age"].fillna(df_train["Age"].median())

df_train["Cabin"] = df_train["Cabin"].fillna("Unknown")

df_train["Embarked"] = df_train["Embarked"].fillna("S")

# testing data

df_test["Age"] = df_test["Age"].fillna(df_test["Age"].median())

df_test["Cabin"] = df_test["Cabin"].fillna("Unknown")

df_test["Embarked"] = df_test["Embarked"].fillna("S")
```

```python
# removing the columns

# For Training
df_train.drop("Name", axis=1, inplace=True)
df_train.drop("Ticket", axis=1, inplace=True)

# For Testing
df_test.drop("Name", axis=1, inplace=True)
df_test.drop("Ticket", axis=1, inplace=True)
```

```python
# removing Cabin columns as well

# For Training
df_train.drop(['Cabin'], axis = 1, inplace = True)

# For Testing
df_test.drop(['Cabin'], axis = 1, inplace = True)
```

```python
# mapping the string to number form

# For Training
df_train["Sex"] = df_train["Sex"].map({"female": 0, "male": 1})
df_train["Embarked"] = df_train["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# For Testing
df_test["Sex"] = df_test["Sex"].map({"female": 0, "male": 1})
df_test["Embarked"] = df_test["Embarked"].map({"S": 0, "C": 1, "Q": 2})
```

```python
# Extrating Features and Target from train data

features = df_train.iloc[0:, 0:-1]
target = df_train.iloc[0:, -1]

print(features)
print(df_test)
```

```python
# training the data using logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

classifier = LogisticRegression()

paramter ={'penalty':['l1','l2','elasticnet'],'C':[0.1,1,10,20,30,40,50,60,100], 'max_iter':[100,110,120,150,200]}

# create a gridsearch object

classifier_model = GridSearchCV(classifier,param_grid=paramter, scoring='accuracy', cv=5)
```

```python
# Training or Fiting The Data

classifier_model.fit(features, target)
```

```python
# checking the best score

print(classifier_model.best_score_ * 100)
```

```python
print(features.columns)
print(df_test.columns)
```

```python
# prediction the Survival rate for testing data

Survival_predicted = classifier_model.predict(df_test)

print(Survival_predicted)
```

```python
# rechanging the data to 

# For Testing
df_test["Sex"] = df_test["Sex"].map({0: "female", 1: "male"})
df_test["Embarked"] = df_test["Embarked"].map({0: "S",1: "C",2: "Q"})
```

```python
# Saving the survivor_predicted column to test.csv

df_test['survivor_predicted'] = Survival_predicted

# saving the csv file

df_test.to_csv("predicted.csv", index=False)
```
